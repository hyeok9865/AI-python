```python
# 과일 이름 리스트
fruits = ["apple", "banana", "cherry", "mango", "orange"]

# 1) for문으로 과일 이름 하나씩 출력
for f in fruits:
    print(f)

# 2) for문으로 새 리스트 만들기: 과일 이름을 대문자로 변환
upper_fruits = []
for f in fruits:
    upper_fruits.append(f.upper())

print("upper_fruits:", upper_fruits)

# 3) for문으로 새 리스트 만들기: 과일 이름 길이(글자 수) 리스트
fruit_lengths = []
for f in fruits:
    fruit_lengths.append(len(f))

print("fruit_lengths:", fruit_lengths)
```

```python
for i, f in enumerate(fruits, start=1):
    print(f"{i}. {f}")
```

```python
# 조선 왕(1대~27대) 이름 리스트
joseon_kings = [
    "태조", "정종", "태종", "세종", "문종", "단종", "세조", "예종", "성종",
    "연산군", "중종", "인종", "명종", "선조", "광해군", "인조", "효종",
    "현종", "숙종", "경종", "영조", "정조", "순조", "헌종", "철종",
    "고종", "순종"
]
```

```python
#몇대왕 누구 인지 나오게, 숫자를 넣으면 왕의 이름이 나오도록

n=int(input("몇 대 왕을 찾을까요?"))
if 1<=n<=len(joseon_kings):
  print(f"{n}대 왕은 {joseon_kings[n-1]}입니다.")
else:
  print("범위를 벗어났습니다. 1~27 사이 숫자를 입력하세요.")
```

```python
#왕의 이름을 넣으면 몇 대 왕인지 나오게, 없으면 없다 라고 나오게.

name=input("왕의 이름을 입력하세요: ")
if name in joseon_kings:
  n=joseon_kings.index(name)+1
  print(f"{name}은(는) 조선 {n}대 왕입니다.")
else:
  print("리스트에 없는 이름입니다.")
```

```python
#조선의 왕을 역순으로 출력해보기

for i in list(reversed(joseon_kings)):
  print(i)
```

```python
#또 다른 방법

for i in range(len(joseon_kings)-1,-1,-1):
  print(joseon_kings[i])
```

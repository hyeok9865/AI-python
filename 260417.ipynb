```python
print("м–ҙм„ңмҳӨм„ёмҡ”! м№ҙнҺҳ мЈјл¬ёмқ„ лҸ„мҷҖл“ңлҰҙкІҢмҡ”.")
print("л©”лүҙ:", ", ".join([f"{name}({price}мӣҗ)" for name, price in cafe_menu.items()]))

# мЈјл¬ё(л©”лүҙлӘ…: мҲҳлҹү)
orders = {}

while True:
    menu_name = input("\nмЈјл¬ён•  л©”лүҙлҘј мһ…л Ҙн•ҳм„ёмҡ”(лҒқ=done): ").strip()

    if menu_name.lower() == "done":
        break

    if menu_name not in cafe_menu:
        print("м—ҶлҠ” л©”лүҙмһ…лӢҲлӢӨ. лӢӨмӢң мһ…л Ҙн•ҙмЈјм„ёмҡ”.")
        continue

    # мҳҲмҷёмІҳлҰ¬
    qty_str = input("мҲҳлҹүмқ„ мһ…л Ҙн•ҳм„ёмҡ”: ").strip()
    if not qty_str.isdigit() or int(qty_str) <= 0:
        print("мҲҳлҹүмқҖ 1 мқҙмғҒмқҳ м •мҲҳлЎң мһ…л Ҙн•ҙмЈјм„ёмҡ”.")
        continue

    qty = int(qty_str)

    # к°ҷмқҖ л©”лүҙлҘј м—¬лҹ¬ лІҲ мЈјл¬ён•  мҲҳ мһҲмңјлӢҲ лҲ„м Ғ
    orders[menu_name] = orders.get(menu_name, 0) + qty
    print(f"нҷ•мқё: {menu_name} {qty}мһ” м¶”к°Җ! (нҳ„мһ¬ {menu_name} мҙқ {orders[menu_name]}мһ”)")

# м „мІҙ кІ°м ң кёҲм•Ў кі„мӮ°(н•өмӢ¬)
total_price = 0
for name, qty in orders.items():
    total_price += cafe_menu[name] * qty

print("\n===== мЈјл¬ё лӮҙм—ӯ =====")
for name, qty in orders.items():
    print(f"- {name} x {qty} = {cafe_menu[name] * qty}мӣҗ")

print("---------------------")
print(f"м „мІҙ кі„мӮ°н•ҙм•ј н•ҳлҠ” кёҲм•Ў(мҙқн•©): {total_price}мӣҗ")
```

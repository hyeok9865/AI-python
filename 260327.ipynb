```python
import random

dice1=random.randint(1,6)
dice2=random.randint(1,6)

print(f'주사위1:{dice1}')
print(f'주사위2:{dice2}')

if dice1>dice2:
  print("주사위1이 더 크다")
elif dice1<dice2:
  print("주사위2가 더 크다")
else:
  print("같다")
```
